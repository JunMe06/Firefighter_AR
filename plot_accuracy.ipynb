```python
#pulic data
import matplotlib.pyplot as plt

window_size = ['5 s','8 s','10 s','12 s','15 s']
gbdt = [92.5, 91.1, 92.6, 92.3, 92]
dt = [89.4, 89.6, 88, 86.7, 91.4]
svm = [92, 92, 92, 93, 92]
rf = [94.2, 94.4, 95.3, 95, 94.9]
knn = [90, 90, 90, 90, 90]

plt.figure(figsize=(10, 6))
plt.plot(window_size,gbdt, 'o--', color='red',  lw = 3,label='GBDT')

plt.plot(window_size,dt, 'o-.', color='blueviolet',lw = 3, label='DT')

plt.plot(window_size,svm, marker='o', linestyle=':', color='blue',lw = 3, label='SVM')

plt.plot(window_size,rf, marker='o', linestyle='-', color='forestgreen',lw = 3, label='RF')

plt.plot(window_size,knn, marker='o', linestyle='-', color='darkorange',lw = 3, label='KNN')

plt.legend(loc='best')
plt.ylim([80, 100])
plt.xlabel('Window size',fontsize=15 )
plt.ylabel('Accuracy (%)', fontsize=15)
plt.title('Accuracy', fontsize=20)

plt.grid()
plt.show()
#plt.gca()
```
